```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/workspaces/codespaces-jupyter/notebooks/hotel_bookings.csv')
```

```python
data = df.copy()
```

```python
plt.figure(figsize = (11 , 8))
sns.set(style= "darkgrid")
plt.title("Total Customers - Monthly", fontdict= {'fontsize': 20})
ax = sns.countplot(x = "arrival_date_month", hue = 'hotel', data= data)
```

En este grafico se observa un incremento de clientes a medida que se acerca el verano

```python
plt.figure(figsize = (15,10))
sns.boxplot(x = "market_segment", y = "stays_in_week_nights", data = data, hue = "hotel", palette = 'Set1');
```

Distribucin de pernoctaciones en hoteles por segmento de mercado y tipo de hotel

```python
(sns.FacetGrid(data, hue = 'is_canceled',
             height = 6,
             xlim = (0,500))
    .map(sns.kdeplot, 'lead_time', shade = True)
    .add_legend());
```
